```python
import pandas as pd
import seaborn as sns

data = pd.read_excel(' ')
print(data)
# groupby() function is used to split the data into groups based on some criteria
prior = data.groupby('class: buys_computer').size().div(len(data))

likelihood = {}
likelihood['Credit_rating'] = data.groupby(['class: buys_computer', 'Credit_rating']).size().div(len(data))/(prior)
likelihood['Age'] = data.groupby(['class: buys_computer', 'Age']).size().div(len(data))/(prior)
likelihood['Income'] = data.groupby(['class: buys_computer', 'Income']).size().div(len(data))/(prior)
likelihood['Student'] = data.groupby(['class: buys_computer', 'Student']).size().div(len(data))/(prior)

# Probability that the person will NOT buy
p_no = likelihood['Age']['no']['youth'] * likelihood['Income']['no']['medium'] * likelihood['Student']['no']['yes'] * likelihood['Credit_rating']['no']['fair'] * prior['no']

# Probability that the person will buy
p_yes = likelihood['Age']['yes']['youth'] * likelihood['Income']['yes']['medium'] * likelihood['Student']['yes']['yes'] * likelihood['Credit_rating']['yes']['fair'] * prior['yes']

print("\n******************************************************************")

print ('Yes (i.e. Probability that the person will buy the computer) --> : ', p_yes)
print ('No (i.e. Probability that the person will not buy the computer) -->  ', p_no)

print("\n******************************************************************")

if p_yes > p_no:
  print("THE PERSON WILL BUY THE COMPUTER...!!")
else:
  print("THE PERSON WILL NOT BUY THE COMPUTER...!!")
print("\n******************************************************************")
```
