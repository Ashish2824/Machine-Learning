```python
from IPython.core.display import display_html
import pandas as pd
import math

df = pd.DataFrame()
df = pd.read_csv('/content/knn.csv')
print(df)

print("\n************************************************************************\n")
df.info()

print("\n\n**********************************************************************************************************************************\n")

distance = []

for i in range(5):
    distance.append(math.sqrt(pow(int(df.iloc[i,1])-int(df.iloc[5,1]),2) + pow(int(df.iloc[i,2])-int(df.iloc[5,2]),2) + pow(int(df.iloc[i,3])-int(df.iloc[5,3]),2) ))

print("Euclidean distances are  --> ", distance)

sort = sorted(distance)
print("\nSorted Euclidean distances are  --> ", sort)

print("\n**********************************************************************************************************************************\n")

t1 = distance.index(sort[0])
t2 = distance.index(sort[1])
t3 = distance.index(sort[2])

print(str(t1) + " " + str(t2) + " " + str(t3) )


```
